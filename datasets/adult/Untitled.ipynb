```python
#adult dataset preprocessing

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('adult.data', sep=",", header=None)
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum().sum()
```

```python
dict = {}
cols = ["feat_" + str(i) for i in df.columns]
# # cols=['radius' ,'texture','perimeter','area',
#       'smoothness' ,'compactness' ,'concavity',
#       'concave_points' , 'symmetry' ,'fractal_dimension', "Diagnosis"]
for i, col in zip(df.columns, cols):
    dict[i] = col

df = df.rename(dict, axis = "columns")
```

```python
df.head()
```

```python
df.rename(columns = {"feat_14":"target"}, inplace = True)
```

```python
labelencoder = LabelEncoder()
df['target'] = labelencoder.fit_transform(df["target"])
```

```python
df.to_csv("adult.csv", index = False)
```
